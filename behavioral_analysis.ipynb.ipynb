```python
import os
import pandas as pd

behavior_files = []

for root, dirs, files in os.walk(dataset_path):
    for file in files:
        if file.endswith(".tsv") and "beh" in root:
            behavior_files.append(os.path.join(root, file))

print("Total behavioral files:", len(behavior_files))
behavior_files[:5]
```

```python
all_data = []

for file in behavior_files:
    df = pd.read_csv(file, sep="\t")
    df["source_file"] = file
    all_data.append(df)

behavior_data = pd.concat(all_data, ignore_index=True)

behavior_data.head()
```
